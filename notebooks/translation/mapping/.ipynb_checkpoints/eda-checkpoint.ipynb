### 2.2. Neighborhood analyses

As a next step, we will compare the neighborhoods of the shared targets in both spaces to better understand the mapping between the two.

#### 2.2.a. Shared neighborhoods

First, we identify the nearest neighbors for each shared target in the strucutral space given by the mean embeddings and the regulatory space

```python
mean_struct_nn_dict = get_sample_neighbor_dict(
    mean_struct_embs["fold_0"], selection=shared_targets
)
```

```python
reg_nn_dict = get_sample_neighbor_dict(reg_embs, selection=shared_targets)
```

To get an idea of the differences in the neighborhoods for the shared targets, we will plot the Jaccard distance of the two sets for increasing k.

```python
def get_iou_neighborhood_for_target(struct_nns, reg_nns, figsize=[6, 4]):
    ious = []
    ks = []
    for k in range(1, len(struct_nns)):
        ks.append(k)
        struct_nn_set = set(list(struct_nns[1 : k + 1]))
        reg_nn_set = set(list(reg_nns[1 : k + 1]))
        ious.append(
            len(struct_nn_set.intersection(reg_nn_set))
            / len(struct_nn_set.union(reg_nn_set))
        )
    return ious, ks
```

```python
cmap = mpl.cm.get_cmap("tab20")
all_ious = []
all_ks = []
for i in range(len(shared_targets)):
    target = sorted(shared_targets)[i]
    if i % 11 == 0:
        fig, ax = plt.subplots(figsize=[8, 6])
        ax.set_ylim([0, 1])
        ax.set_xlabel("k-nearest neighbors")
        ax.set_ylabel("IoU")
    ious, ks = get_iou_neighborhood_for_target(
        mean_struct_nn_dict[target], reg_nn_dict[target]
    )
    ax.plot(ks, ious, label=target, c=cmap((i % 11) / 10))
    all_ious.append(ious)
    all_ks.append(ks)
    ax.legend()
plt.show()
plt.close()
```

```python
def get_random_iou(struct_embs, reg_embs, selection, b=1000, random_state=1234):
    np.random.seed(random_state)
    perm_ks = []
    perm_ious = []
    for i in tqdm(range(b)):
        struct_perm_embs = struct_embs.copy()
        struct_perm_embs.index = np.random.permutation(list(struct_perm_embs.index))

        reg_perm_embs = reg_embs.copy()
        reg_perm_embs.index = np.random.permutation(list(reg_perm_embs.index))

        struct_nn_dict = get_sample_neighbor_dict(struct_perm_embs, selection=selection)
        reg_nn_dict = get_sample_neighbor_dict(reg_perm_embs, selection=selection)

        all_ious = []
        all_ks = []
        for target in sorted(selection):
            ious, ks = get_iou_neighborhood_for_target(
                mean_struct_nn_dict[target], reg_nn_dict[target]
            )
            all_ious.append(ious)
            all_ks.append(ks)
        perm_ks.append(np.array(all_ks).mean(axis=0))
        perm_ious.append(np.array(all_ious).mean(axis=0))
    return np.array(perm_ious), np.array(perm_ks)
```

```python
random_ious, random_ks = get_random_iou(
    mean_struct_embs["fold_0"], reg_embs, selection=shared_targets
)
```

```python
mpl.style.use("ggplot")
fig, ax = plt.subplots(figsize=[9, 6])
ax.plot(
    np.array(all_ks).mean(axis=0),
    np.array(all_ious).mean(axis=0),
    label="observed average",
    c="blue",
)
random_baseline = pd.DataFrame(
    {"k": np.array(list(random_ks)).ravel(), "iou": np.array(list(random_ious)).ravel()}
)
ax = sns.lineplot(data=random_baseline, x="k", y="iou", ax=ax, ci="sd", label="random")
ax.set_ylim([0, 1])
ax.set_title("IoU of physical and regulatory neighborhood")
ax.set_xlabel("k-nearest neighbors")
ax.set_ylabel("IoU")
plt.show()
```
