# Cluster analyses of the morphological profiles


To understand potential short comings of our inference method of the morphological profiles, we will cluster the individual gene targets of the set of structural regulators using their corresponding morphological profiles as computed by the authors.

```python
def get_emd_for_embs(embs, label_col, metric="euclidean"):
    targets = np.unique(embs.loc[:, label_col])
    n_targets = len(targets)
    wd_mtx = np.infty * np.ones((n_targets, n_targets))
    for i in tqdm(range(n_targets), desc="Compute EMD"):
        source = targets[i]
        xs = np.array(embs.loc[embs.loc[:, label_col] == source]._get_numeric_data())
        ns = len(xs)
        ps = np.ones((ns,)) / ns
        for j in range(i, n_targets):
            target = targets[j]
            if source == target:
                wd_st = 0
            else:
                xt = np.array(
                    embs.loc[embs.loc[:, label_col] == target]._get_numeric_data()
                )
                nt = len(xt)
                pt = np.ones((nt,)) / nt
                m = ot.dist(xs, xt, metric=metric)
                m = m / m.max()
                wd_st = ot.emd2(ps, pt, m, numItermax=1e9)
            wd_mtx[i, j] = wd_st
            wd_mtx[j, i] = wd_st
    wd_df = pd.DataFrame(wd_mtx, columns=list(targets), index=list(targets))
    return wd_df
```

## 2. Clustering

### 2.1. Mean clustering

```python
mean_profiles = morph_profiles.groupby("labels").median().drop(columns=["ImageNumber"])
# mean_profiles = mean_profiles.drop(columns=["Nuclei_Number_Object_Number"])
```

```python
mean_profiles = pd.DataFrame(
    StandardScaler().fit_transform(mean_profiles),
    index=mean_profiles.index,
    columns=mean_profiles.columns,
)
```

```python
cluster_labels = np.array(reg_embs_clusters.loc[mean_profiles.index]).ravel()
```

```python
mpl.style.use("default")
fig, ax = plt.subplots(figsize=[12, 9])
ax = plot_tsne_embs(
    mean_profiles,
    ax=ax,
    perplexity=8,
    random_state=1234,
    hue=np.array(cluster_labels).astype(str),
    hue_order=[
        "ECM interactions",
        "Cell cycle control",
        "Signal transduction",
        "Cytoskeletal organization",
        "DNA damage response",
    ],
    palette={
        "ECM interactions": "tab:blue",
        "Cell cycle control": "tab:orange",
        "Signal transduction": "tab:green",
        "Cytoskeletal organization": "tab:red",
        "DNA damage response": "tab:purple",
    },
)
ax.set_title("Morpholigcal profile mean embeddings")
ax.legend(title="Cluster", loc="upper left")
# ax.set_xlim([-40, 50])
plt.show()
```

### 2.2. Wasserstein distance clustering

```python
labels = np.array(morph_profiles.labels)
morph_profiles = pd.DataFrame(
    StandardScaler().fit_transform(morph_profiles._get_numeric_data())
)
morph_profiles["labels"] = labels
sampled_morph_profiles = morph_profiles.groupby("labels", group_keys=False).apply(
    lambda x: x.sample(frac=0.25)
)
wdist_profiles = get_emd_for_embs(embs=sampled_morph_profiles, label_col="labels")
```

```python
linkage = hc.linkage(squareform(wdist_profiles), method="average")
ax = sns.clustermap(
    wdist_profiles,
    row_linkage=linkage,
    col_linkage=linkage,
    figsize=[12, 12],
    cmap="seismic",
)
print("Wasserstein distance matrix of the morphological embedding space")
plt.show()
```
