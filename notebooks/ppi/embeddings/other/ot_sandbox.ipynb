```python
from ot import plot as ot_plt
```

```python
embs = pd.read_hdf(
    "../../../data/experiments/image_embeddings/specificity_target_emb_cv_strat/final_1024/fold_0/test_latents.h5",
    index_col=0,
)
label_dict = {v: k for k, v in label_dict.items()}
embs.iloc[:, -1] = embs.iloc[:, -1].map(label_dict)
embs.head()
```

```python
xs = np.array(embs.loc[embs.labels == "JUN"].iloc[:, :-1])
xt = np.array(embs.loc[embs.labels == "EMPTY"].iloc[:, :-1])
pc = PCA(n_components=2)
transformed = pc.fit_transform(np.concatenate((xs, xt)))
xs = transformed[: len(xs)]
xt = transformed[len(xs) :]
```

```python
ns = len(xs)
nt = len(xt)

a = np.ones((ns,)) / ns
b = np.ones((nt,)) / nt

M = ot.dist(xs, xt)
M = M / M.max()
```

```python
wd = ot.emd(a, b, M, numItermax=1e8)
```

```python
plt.figure(1)
plt.plot(xs[:, 0], xs[:, 1], "+b", label="JUN")
plt.plot(xt[:, 0], xt[:, 1], "xr", label="CTRL")
plt.legend(loc=0)
plt.title("Source and target distributions")

plt.figure(2)
plt.imshow(M, interpolation="nearest")
plt.title("Cost matrix M")
```

```python
plt.figure(3)
plt.imshow(wd, interpolation="nearest")
plt.title("OT matrix G0")

plt.figure(4, figsize=[24, 16])
ot_plt.plot2D_samples_mat(xs, xt, wd, c=[0.5, 0.5, 1])
plt.plot(xs[:, 0], xs[:, 1], "+b", label="Source samples")
plt.plot(xt[:, 0], xt[:, 1], "xr", label="Target samples")
plt.legend(loc=0)
plt.title("OT matrix with samples")
```
